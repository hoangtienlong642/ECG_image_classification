```python
from helper_code import *
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from upsetplot import UpSet


```

```python


def load_data(data_folder):
    records = find_records(data_folder)
    num_records = len(records)
    classification_labels = list()
    for i in range(num_records):
        record = os.path.join(data_folder, records[i])
        labels = load_labels(record)
        if any(label for label in labels):
            classification_labels.append(labels)
    classes = sorted(set.union(*map(set, classification_labels)))
    classification_labels = compute_one_hot_encoding(classification_labels, classes)
    df = pd.DataFrame(records, columns=['record'])
    labels_df = pd.DataFrame(classification_labels, columns=classes)
    # df = pd.concat([df, labels_df], axis=1)
    return df, labels_df
```

```python


labels_df = load_data('data/train_data2')[1]
plt.figure(figsize=(10, 6))
sns.heatmap(labels_df.T, annot=True, cmap='viridis', cbar=False, xticklabels=False)
plt.title("Heatmap")
plt.ylabel("Labels")
plt.xlabel("Samples")
plt.show()
```

```python
label_counts = labels_df.sum(axis=0)
plt.figure(figsize=(8, 4))
sns.barplot(x=label_counts.index, y=label_counts.values, palette='viridis')
plt.title("Label Frequency")
plt.xlabel("Labels")
plt.ylabel("Count")
plt.show()
```

```python
df_labels_upset = labels_df.astype(bool)
upset_data = df_labels_upset.value_counts()
upset = UpSet(upset_data, subset_size='count', sort_by='degree')
upset.plot()
plt.suptitle("UpSet Plot of Multilabel Data")
plt.show()
```
